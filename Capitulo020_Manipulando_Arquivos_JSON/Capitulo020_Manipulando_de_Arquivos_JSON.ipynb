#### Diferenças entre um arquivo JSON e um dicionário Python

Pode ser tentador acreditar que a principal diferença entre um arquivo JSON e um dicionário Python seja somente a maneira como usamos as aspas. No entanto, a distinção entre os dois vai além disso. A seguir, apresento algumas diferenças chave:

1. Natureza e Uso:

- JSON (JavaScript Object Notation): É um formato de intercâmbio de dados leve, de fácil leitura e escrita para humanos e fácil de ser analisado e gerado por máquinas. É independente de linguagem, o que significa que dados podem ser trocados entre diferentes linguagens de programação usando JSON.
- Dicionário (dict em Python): É uma estrutura de dados específica da linguagem Python que armazena pares de chave-valor.

2. Representação:
- JSON: É uma string. Exemplo: '{"nome": "João", "idade": 30}'
- Dicionário: É um objeto Python. Exemplo: {"nome": "João", "idade": 30}

3. Aspas:
- JSON: Sempre utiliza aspas duplas (") para representar chaves e strings.
- Dicionário: Pode utilizar aspas simples (') ou aspas duplas (").

4. Tipos de Dados:
- JSON: Tem tipos de dados limitados: número, string, booleano, array, objeto e null.
- Dicionário: Pode armazenar quase qualquer tipo de dado Python como valor: números, strings, listas, outros dicionários, objetos, funções e assim por diante.

5. Função:
- JSON: Serve principalmente para transmitir dados entre um servidor e uma aplicação web ou entre diferentes partes de uma aplicação.
- Dicionário: Serve como uma estrutura de dados fundamental em Python para armazenar e organizar dados.

Para converter entre essas duas representações em Python, usamos o módulo json:

- json.dumps(dict_obj) converte um dicionário em uma string JSON.
- json.loads(json_str) converte uma string JSON em um dicionário.

#### *Exemplo 1: Criando e salvando um arquivo JSON*

```python
import json

# Dicionário contendo dados profissionais
dict_prof1={"Professor": "Allan",
  "Área": "Matemática",
  "Ano": 2023,
  "Semeste": 1,
  "Turmas": 3,
  "Matérias": [
      "Matemática Aplicada",
      "Matemática Discreta I",
      "Matemática Discreta II"
      ]
}
```

```python
# Criando um arquivo Json
with open('dados_prof.json','w') as arquivo:
    arquivo.write(json.dumps(dict_prof1))
```

Após a execução da célula acima foi criado o arquivo dados_prof.json. Vá até a pasta deste arquivo e abra-o com bloco de notas.

```python
# Leitura do arquivo
with open('dados_prof.json','r') as arquivo:
    texto = arquivo.read()
    dados_prof = json.loads(texto)
```

##### Exemplo 1: Gerando listas contendo paises e suas respectivas populações no ano de 2023

```python
# Importando o módulo JSON
import json

# Leitura do arquivo
with open('countries-table.json','r') as arquivo:
    texto = arquivo.read()
    dados_paises = json.loads(texto)
```

```python
# Cria uma lista vazia para armazenar os países
lista_paises = []  

# Inicializa a variável de controle do loop
i = 0  

# Itera sobre cada item em dados_paises
for item in dados_paises:  
    
    # Adiciona o nome do país à lista_paises
    lista_paises.append(dados_paises[i]['country'])  
    
    # Incrementa a variável de controle do loop
    i = i + 1  
```

```python
# Lista vazia com populações no ano de 2023
lista_pop2023 = []

# Contador
i=0

# Adiciona as populações dos países da posição i
for item in dados_paises:
    lista_pop2023.append(dados_paises[i]['pop2023'])
    i=i+1
```

##### Exemplo 2: Gerando listas contendo paises e suas respectivas densidades no ano de 2023

```python
# Importando o módulo JSON
import json

# Leitura do arquivo
with open('countries-table.json','r') as arquivo:
    texto = arquivo.read()
    dados_paises = json.loads(texto)
    
# Cria uma lista vazia para armazenar os países
lista_paises = []  

# Cria uma lista vazia para armazenar as densidades
lista_den2023 = []

# Inicializa a variável de controle do loop
i = 0  

# Itera sobre cada item em dados_paises
for item in dados_paises:  
    
    # Adiciona o nome do país à lista_paises
    lista_paises.append(dados_paises[i]['country'])  
    
    # Adiciona a densidade do país à lista_densidades
    lista_den2023.append(dados_paises[i]['density']) 
    
    # Incrementa a variável de controle do loop
    i = i + 1  

# Criando lista com países e respectivas populações no ano de 2023
paises_den2023 = list(zip(lista_paises, lista_den2023))
```

##### Coeficiente de determinação

O coeficiente de determinação, representado por $\rho^2$, varia de 0 a 1 e possui a seguinte interpretação:

- Um valor próximo de 1 indica um ajuste mais preciso do modelo aos dados e uma maior capacidade do modelo de explicar a variação observada nos dados.

- Um valor de coeficiente de determinação igual a 0 indica que o modelo não consegue explicar a variação dos dados e a relação entre as variáveis não pode ser representada adequadamente pela reta de regressão.

O coeficiente de determinação é uma medida importante na avaliação da qualidade do ajuste do modelo aos dados, pois nos fornece uma indicação de quanto da variação total dos dados é explicada pelo modelo.

##### Exemplo: Calculando o coeficiente de determinação e o ajuste linear entre as populações de 2023 e 2030

```python
# Importando o módulo JSON
import json

# Leitura do arquivo
with open('countries-table.json','r') as arquivo:
    texto = arquivo.read()
    dados_paises = json.loads(texto)
    
# Cria uma lista vazia para armazenar os países
lista_pop2023 = []  

# Cria uma lista vazia para armazenar as densidades
lista_pop2030 = []

# Inicializa a variável de controle do loop
i = 0  

# Itera sobre cada item em dados_paises
for item in dados_paises:  
    
    # Adiciona o nome do país à lista_paises
    lista_pop2023.append(dados_paises[i]['pop2023'])  
    
    # Adiciona a densidade do país à lista_densidades
    lista_pop2030.append(dados_paises[i]['pop2030']) 
    
    # Incrementa a variável de controle do loop
    i = i + 1  
```

```python
# Criando lista com países e respectivas populações no ano de 2023
lista_pop2023_2030 = list(zip(lista_pop2023, lista_pop2030))

# Exibindo a lista de populações em 2023 e 2030
lista_pop2023_2030
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Entradas
x = lista_pop2023
y = lista_pop2030

# Converter as listas x e y para números float
x = np.array(x, dtype=float)
y = np.array(y, dtype=float)

# Ajuste linear
coefficients_linear = np.polyfit(x, y, 1)
correlation_linear = np.corrcoef(x, y)[0, 1]
equation_linear = f"lin(x) = {coefficients_linear[0]:.10f}x + {coefficients_linear[1]:.10f}"
ss_total_linear = np.sum((y - np.mean(y)) ** 2)
ss_residual_linear = np.sum((y - (coefficients_linear[0] * x + coefficients_linear[1])) ** 2)
r_squared_linear = 1 - (ss_residual_linear / ss_total_linear)

# Plot da reta de regressão
plt.scatter(x, y, color='b', label='Dados')
plt.plot(x, coefficients_linear[0] * x + coefficients_linear[1], color='r', label='Regressão Linear')
plt.xlabel('População em 2023')
plt.ylabel('População em 2030')
plt.title('Reta de Regressão Linear')
plt.legend()
plt.show()

# Exibindo os resultados
print("Ajuste linear:")
print("Equação:", equation_linear)
print("Coeficiente de Determinação:", r_squared_linear)
```

```python
import requests
import geopandas as gpd
import matplotlib.pyplot as plt

# Baixando os dados
url = Link3
response = requests.get(url)
data = response.json()

# Convertendo para GeoDataFrame
gdf = gpd.GeoDataFrame.from_features(data["features"])

# Filtrando apenas Brazil, Afghanistan e Angola
countries_to_plot = ["Brazil","Afghanistan", "Angola", "Portugal"]
gdf_filtered = gdf[gdf['name'].isin(countries_to_plot)]

# Cores para cada país
colors = {
    "Brazil": "green",
    "Afghanistan": "blue",
    "Angola": "red",
    "Portugal": "black"
}

# Plotando cada país com sua cor correspondente
fig, ax = plt.subplots()
for country in countries_to_plot:
    gdf_filtered[gdf_filtered['name'] == country].plot(ax=ax, color=colors[country])

plt.title("Countries")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

#### *Extração a partir de arquivo local*

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import json

# Caminho para o arquivo JSON baixado
file_path = 'world-countries.json'

# Carregando o arquivo JSON
with open(file_path) as file:
    data = json.load(file)

# Convertendo para GeoDataFrame
gdf = gpd.GeoDataFrame.from_features(data["features"])

# Filtrando apenas Brazil, Afghanistan e Angola
countries_to_plot = ["Brazil", "Afghanistan", "Angola"]
gdf_filtered = gdf[gdf['name'].isin(countries_to_plot)]

# Cores para cada país
colors = {
    "Brazil": "green",
    "Afghanistan": "blue",
    "Angola": "red"
}

# Plotando cada país com sua cor correspondente
fig, ax = plt.subplots()
for country in countries_to_plot:
    gdf_filtered[gdf_filtered['name'] == country].plot(ax=ax, color=colors[country])

plt.title("Countries")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
